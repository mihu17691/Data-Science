```python
# Convert data into percentage format for better analysis

returns = np.log(df).diff()
returns.head()
```

```python
# LB and AC relationship using scatter plot

from matplotlib import rcParams
rcParams['figure.figsize'] = 8,6
plt.axvline(0, c=(.5, .5, .5), ls='--')
plt.axhline(0, c=(.5, .5, .5), ls='--')
plt.style.use('seaborn')
plt.scatter(returns.LB, returns.AC, c = returns.AC, cmap = 'Spectral', edgecolor = 'k')
plt.colorbar(orientation='horizontal')
```
